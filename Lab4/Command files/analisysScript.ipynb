```python
import pandas as pd
import numpy as np
import time
from datetime import datetime, date, time, timedelta
```

```python
data = pd.read_csv("../Analysis Data/analysisData_1.csv")
data
```

```python
weeks_in_charts = {}


for d,r in data.iterrows():
    total_weeks = r["week"]  
    weeks_in_charts[r["track"]] = total_weeks
    
lines = ["#Analisys of processed data","","Every song will have it's total time in top charts listed",""]
for k, v in weeks_in_charts.items():
    lines.append("\" {} \" was in charts for: {} weeks\n".format(k,v))
    
with open("../Docs/Analisys.md",'w') as file:
    for line in lines:
        file.write(line)
```
