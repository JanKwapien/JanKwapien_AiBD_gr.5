```python
import pandas as pd
import numpy as np
import time
from datetime import datetime, date, time, timedelta
```

```python
data = pd.read_csv("../Original Data/billboard.csv")
data = data.drop_duplicates()
data
```

## Processing
Selecting "year","artist", "track" and "time" from original data, adding "date","week" and "place" fields.

Each row will respresent a song every week with it's rank.

```python
print(len(data))
songs = data.loc[:,['year','artist.inverted','track','time','genre']]
places = data.loc[:,'x1st.week':]

data_place_by_week = pd.DataFrame(columns = ["year","artist", "track","time","date","week","place"])
for d,r in data.iterrows():
    week = 0
    for w, place in places.loc[d,:].iteritems():
        
        if not np.isnan(place):
            date = datetime.strptime(r['date.entered'], '%Y-%m-%d') + timedelta(weeks = week)
            date = date.strftime("%Y-%m-%d")
            #print(date.strftime("%Y-%m-%d"),weeks+1,place)
            new_row = [r['year'],r['artist.inverted'],r['track'],r['time'],date ,week + 1,int(place)]
            #print(new_row)
            data_place_by_week.loc[len(data_place_by_week)]= new_row
        week+=1
        
    total_weeks = week

data_place_by_week.to_csv("../Analysis Data/analysisData_1.csv")

data_place_by_week
```
